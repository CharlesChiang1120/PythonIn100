```python
import random
from replit import clear
```

```python
stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']
```

```python
with open('vocabulary.txt', 'r') as file:
    word_list = file.read().replace('\n', '')
```

```python
word_list = word_list.replace("\'", "").split(",")
```

```python
answer = random.choice(word_list)
word_length = len(answer)
blank_list = list('_'*word_length)
```

```python
lives = 6
end_of_game = False
while end_of_game != True:
    if "_" in blank_list != True:
        input_str = input('Guess a letter: ').lower()
        clear()
        print("-----------------------")
        if input_str in answer:   
            for position in range(word_length):
                letter = answer[position]
                if input_str == letter:
                    blank_list[position] = letter
            print(" ".join(blank_list))
            print("-----------------------")
        else:
            print(stages[lives])
            print("-----------------------")
            lives -= 1
            if lives == -1:
                end_of_game = True
                print("You Lose")
                print("Answer is: ", answer)
    else:
        end_of_game == True
        print("You Win")
        print("-----------------------")
        break    
```
